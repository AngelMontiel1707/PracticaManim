La función grafvecs se define para aceptar dos argumentos:

vecs: Una lista de vectores que se desean graficar.

cols: Una lista de colores, donde cada color corresponde a uno de los vectores en la lista vecs.



Se crea una nueva figura para el gráfico con un tamaño específico de 5x5 pulgadas. 
Esto proporciona un área donde los vectores serán visualizados.


La función itera sobre la lista de vectores vecs utilizando un bucle

Para cada índice en la lista de vectores, se obtiene el vector correspondiente.

Se dibuja un vector en el gráfico comenzando desde el origen (0,0) y extendiéndose hasta el punto definido por el vector actual.

El color del vector se asigna según la lista cols, de modo que cada vector tiene el color correspondiente de la lista.

```python
def grafvecs(vecs, cols):
    plt.figure(figsize=(5,5))
    # vecs: lista de vectores
    for i in range(len(vecs)):
        vec = vecs[i]
        plt.quiver(0,0, vec[0], vec[1], color = cols[i], 
                   angles='xy', scale_units='xy', scale=1)
```

grafvecs: Define una función para graficar múltiples vectores con colores específicos.
    
plt.figure: Crea una nueva figura de gráfico con tamaño específico.
    
Bucle for: Itera sobre los vectores y los grafica con los colores especificados.

```python
v1 = np.array([1,2])
v2 = np.array([3,4])
v3 = np.array([8,4])

grafvecs([v1, v2, v3], ['blue', 'red', 'black'])
plt.xlim(0, max(v1[0], v2[0], v3[0]))
plt.ylim(0, max(v1[1], v2[1], v3[1]))
```
