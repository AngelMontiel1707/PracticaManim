```python
import matplotlib.pyplot as plt
import numpy as np

# Coordenadas del triángulo

triangulo_x = [1, 1.5, 2, 1]
triangulo_y = [1, 2, 1, 1]


# Función para la traslación

def traslacion(x, y, tx, ty):
    x_trasladado = [i + tx for i in x]
    y_trasladado = [i + ty for i in y]
    return x_trasladado, y_trasladado

# Función para la escalación

def escalacion(x, y, sx, sy):
    x_escalado = [i * sx for i in x]
    y_escalado = [i * sy for i in y]
    return x_escalado, y_escalado

# Función para la rotación

def rotacion(x, y, angulo):
    rad = np.radians(angulo)
    x_rotado = [x[i] * np.cos(rad) - y[i] * np.sin(rad) for i in range(len(x))]
    y_rotado = [x[i] * np.sin(rad) + y[i] * np.cos(rad) for i in range(len(y))]
    return x_rotado, y_rotado

# Trasladamos la figura

triangulo_trasladado = traslacion(triangulo_x, triangulo_y, 1, 1)


# Escalamos la figura 

triangulo_escalado = escalacion(triangulo_x, triangulo_y, 3, 1.5)


# Rotamos la figura a 45 grados

triangulo_rotado = rotacion(triangulo_x, triangulo_y, 45)


# Graficamos todos los triangulos

plt.figure(figsize=(10, 8))

# Agregamos el triangulo original

plt.plot(triangulo_x, triangulo_y, 'yellow', label='Triángulo ')

# Figura trasladada

plt.plot(triangulo_trasladado[0], triangulo_trasladado[1], 'green',label='Triángulo trasladado')


# Figura escalada

plt.plot(triangulo_escalado[0], triangulo_escalado[1], 'red',label='Triángulo escalado')


# Figura rotada

plt.plot(triangulo_rotado[0], triangulo_rotado[1], 'blue',label='Triángulo rotado')

#Agrgamos etiquetas 

plt.legend()
plt.title('Traslación, Escalación y Rotación')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Coordenadas originales del triángulo y el cuadrado

cuadrado_x = [1, 2, 2, 1, 1]
cuadrado_y = [1, 1, 2, 2, 1]

# Función para la traslación
def traslacion(x, y, tx, ty):
    x_trasladado = [i + tx for i in x]
    y_trasladado = [i + ty for i in y]
    return x_trasladado, y_trasladado

# Función para la escalación
def escalacion(x, y, sx, sy):
    x_escalado = [i * sx for i in x]
    y_escalado = [i * sy for i in y]
    return x_escalado, y_escalado

# Función para la rotación
def rotacion(x, y, angulo):
    rad = np.radians(angulo)
    x_rotado = [x[i] * np.cos(rad) - y[i] * np.sin(rad) for i in range(len(x))]
    y_rotado = [x[i] * np.sin(rad) + y[i] * np.cos(rad) for i in range(len(y))]
    return x_rotado, y_rotado

# Trasladar las figuras (por ejemplo, 1 unidad en x y 1 unidad en y)

cuadrado_trasladado = traslacion(cuadrado_x, cuadrado_y, 2, 3)

# Escalar las figuras (por ejemplo, 2x en x y 0.5x en y)

cuadrado_escalado = escalacion(cuadrado_x, cuadrado_y, 3, 1.5)

# Rotar las figuras (por ejemplo, 45 grados)

cuadrado_rotado = rotacion(cuadrado_x, cuadrado_y, 45)

# Graficar todas las figuras
plt.figure(figsize=(12, 10))

# Figuras originales
plt.plot(cuadrado_x, cuadrado_y, 'green', label='Cuadrado original')


# Figuras trasladadas
plt.plot(cuadrado_trasladado[0], cuadrado_trasladado[1], 'red',label='Cuadrado trasladado')

# Figuras escaladas
plt.plot(cuadrado_escalado[0], cuadrado_escalado[1], 'blue',label='Cuadrado escalado')

# Figuras rotadas
plt.plot(cuadrado_rotado[0], cuadrado_rotado[1], 'black', label='Cuadrado rotado')

plt.legend()
plt.title('Transformaciones: Traslación, Escalación y Rotación')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)
plt.axis('equal')
plt.show()
```
