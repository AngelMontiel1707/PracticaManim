Descripción: Esta función construye una parte del polígono convexo utilizando el algoritmo de envolvente convexa, conocido como el método del "giro a la derecha" o "gift wrapping".

Uso de numpy.linalg.det: El determinante se utiliza para verificar si los puntos están en una orientación que indica un giro a la derecha, lo que ayuda a construir la envolvente convexa.

Eliminación de puntos: Si los puntos no cumplen con la condición del giro a la derecha, se eliminan del polígono para garantizar que permanezca convexo.

```python
def convex_hull():
    coord_points.sort()
    l_upper = turn_right()
    coord_points.reverse()
    l_lower = turn_right()
    l = l_upper + l_lower
    return l
```

Descripción: Esta función construye la envolvente convexa completa utilizando las partes superior e inferior del polígono.

Uso de sort() y reverse(): Ordena las coordenadas para simplificar la construcción de la envolvente convexa y luego construye ambas partes, la superior e inferior.

Construcción y graficación: Se genera la envolvente convexa y se grafica utilizando las funciones definidas.

```python
import random as rand
import numpy as np
import matplotlib.pyplot as plt

# -----------------FUNCIONES----------------- #

def turn_right(coord_points):
    array = [coord_points[0], coord_points[1]]
    for i in range(2, len(coord_points)):
        array.append(coord_points[i])
        while len(array) > 2 and np.linalg.det([array[-3], array[-2], array[-1]]) > 0:
            array.pop(-2)
    return array

def convex_hull(coord_points):
    coord_points.sort()
    l_upper = turn_right(coord_points)
    coord_points.reverse()
    l_lower = turn_right(coord_points)
    l = l_upper + l_lower
    return l

def graph(convex_pol1, coord_points1, convex_pol2, coord_points2):
    # Preparando listas adecuadas para graficar en matplotlib
    x_points1 = [i[0] for i in coord_points1]
    y_points1 = [i[1] for i in coord_points1]
    x_polygon1 = [i[0] for i in convex_pol1]
    y_polygon1 = [i[1] for i in convex_pol1]

    x_points2 = [i[0] for i in coord_points2]
    y_points2 = [i[1] for i in coord_points2]
    x_polygon2 = [i[0] for i in convex_pol2]
    y_polygon2 = [i[1] for i in convex_pol2]

    # Definiendo límites extremos de la gráfica
    x_lim_der = max(max(x_points1), max(x_points2)) + 5
    y_lim_sup = max(max(y_points1), max(y_points2)) + 5
    x_lim_izq = min(min(x_points1), min(x_points2)) - 5
    y_lim_inf = min(min(y_points1), min(y_points2)) - 5

    # Asignación de los límites extremos
    plt.xlim(x_lim_izq, x_lim_der)
    plt.ylim(y_lim_inf, y_lim_sup)

    # Graficación
    plt.title('Problema: Envolventes Convexas de Dos Polígonos')
    plt.xlabel('Eje de las abscisas')
    plt.ylabel('Eje de las ordenadas')

    # Graficando el primer conjunto de puntos y su envolvente convexa
    plt.plot(x_points1, y_points1, 'bo')  # Puntos en azul
    plt.plot(x_polygon1, y_polygon1, '-b', linewidth=1.0, label='Polígono 1')  # Línea en azul

    # Graficando el segundo conjunto de puntos y su envolvente convexa
    plt.plot(x_points2, y_points2, 'ro')  # Puntos en rojo
    plt.plot(x_polygon2, y_polygon2, 'r-', linewidth=1.0, label='Polígono 2')  # Línea en rojo

    plt.legend()
    plt.show()

# -------------------------------------------- #

# Generación de coordenadas aleatorias para dos conjuntos de puntos

num_points = 50
coord_points1 = [[rand.uniform(0, 50), rand.uniform(0, 100), 1.0] for _ in range(num_points)]
coord_points2 = [[rand.uniform(50, 100), rand.uniform(0, 100), 1.0] for _ in range(num_points)]

# Creación de los polígonos convexos para cada conjunto de puntos
convex_pol1 = convex_hull(coord_points1)
convex_pol2 = convex_hull(coord_points2)

# Graficación de ambos polígonos
graph(convex_pol1, coord_points1, convex_pol2, coord_points2)
```

Se modificaron las funciones turn_right y convex_hull para que tomen los puntos como argumento, lo que permite usarlas para múltiples conjuntos de puntos.

Se añadieron dos conjuntos de puntos aleatorios, coord_points1 y coord_points2, para generar dos polígonos distintos.

Se actualizó la función graph para recibir y graficar ambos polígonos en la misma figura.
